```python
import requests
import os
import random

# Create a directory to store the downloaded images
os.makedirs("rami_malek_photos", exist_ok=True)

# Initialize variables
base_url = "https://rami-malek.org/photos/albums/Movies/2018%20Bohemian%20Rhapsody/caps/thumb_BohemianRhapsody-"
total_photos = 2805
max_photos = 500
unique_photos = set()
downloaded = 0

# Download unique Rami Malek pictures
while downloaded < max_photos:
    # Generate a random photo index
    photo_index = random.randint(1, total_photos)
    formatted_index = str(photo_index).zfill(4)  # Zero-padding the index

    # Construct the photo URL
    photo_url = f"{base_url}{formatted_index}.jpg"

    # Check if the photo has already been downloaded
    if photo_url not in unique_photos:
        # Download the photo
        response = requests.get(photo_url)

        # Save the photo to the directory
        with open(f"rami_malek_photos/rami_{downloaded + 1}.jpg", "wb") as file:
            file.write(response.content)

        # Update the downloaded photo count and add the URL to the set of unique photos
        downloaded += 1
        unique_photos.add(photo_url)

        print(f"Downloaded photo {downloaded}/{max_photos}: {photo_url}")

print("Download complete!")
```

```python
import requests
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Create a directory to store the downloaded photos
os.makedirs("freddie_mercury_photos", exist_ok=True)

# Base URL of the website to download photos from
base_url = "https://www.theplace.ru/photos/freddie_mercury/"

# Number of pages to loop through
num_pages = 25

# Loop through each page
for page in range(1, num_pages + 1):
    # Generate the URL for the current page
    page_url = base_url + f"?page={page}"

    # Send a GET request to the page
    response = requests.get(page_url)

    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all image tags
    image_tags = soup.find_all("img")

    # Download photos from the current page
    for image_tag in image_tags:
        # Get the source URL of the photo
        photo_url = urljoin(base_url, image_tag["src"])

        # Check if the photo URL ends with ".jpg" to ensure it's an image
        if photo_url.endswith(".jpg"):
            # Extract the filename from the URL
            filename = os.path.basename(photo_url)

            # Send a GET request to download the photo
            photo_response = requests.get(photo_url)

            # Save the photo to the directory
            with open(os.path.join("freddie_mercury_photos", filename), "wb") as file:
                file.write(photo_response.content)

            print(f"Downloaded photo: {filename}")

print("Download complete!")
```

```python
import os
from PIL import Image

# Function to resize an image to the target size
def resize_image(image_path, target_size):
    image = Image.open(image_path)
    resized_image = image.resize(target_size, Image.ANTIALIAS)
    resized_image.save(image_path)

# Directory paths
rami_malek_directory = '/content/rami_malek_photos'
freddie_mercury_directory = '/content/freddie_mercury_photos'

# Target size for resizing
target_size = (256, 256)

# Resize images in the rami_malek_photos directory
for filename in os.listdir(rami_malek_directory):
    if filename.endswith('.jpg'):
        image_path = os.path.join(rami_malek_directory, filename)
        resize_image(image_path, target_size)

# Resize images in the freddie_mercury_photos directory
for filename in os.listdir(freddie_mercury_directory):
    if filename.endswith('.jpg'):
        image_path = os.path.join(freddie_mercury_directory, filename)
        resize_image(image_path, target_size)

print("Image resizing complete!")
```

```python
import os
import pandas as pd

# Directory paths
freddie_mercury_directory = '/content/freddie_mercury_photos'
rami_malek_directory = '/content/rami_malek_photos'

# Create a list to store the file paths and labels
file_paths = []
labels = []

# Load files from the freddie_mercury_photos directory
for filename in os.listdir(freddie_mercury_directory):
    if filename.endswith('.jpg'):
        file_path = os.path.join(freddie_mercury_directory, filename)
        file_paths.append(file_path)
        labels.append(1)  # is_Freddie = 1 for freddie_mercury_photos

# Load files from the rami_malek_photos directory
for filename in os.listdir(rami_malek_directory):
    if filename.endswith('.jpg'):
        file_path = os.path.join(rami_malek_directory, filename)
        file_paths.append(file_path)
        labels.append(0)  # is_Freddie = 0 for rami_malek_photos

# Create the pandas DataFrame
data = {'path': file_paths, 'is_Freddie': labels}
df = pd.DataFrame(data)
```

```python
df['is_Freddie'] = df['is_Freddie'].astype(str)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Set up data generators for image loading and augmentation
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

# Define the input shape for the CNN model
input_shape = (256, 256, 3)

# Build the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Set up batch size and number of epochs for training
batch_size = 32
epochs = 10

# Set up data generators for training and testing
train_generator = train_datagen.flow_from_dataframe(dataframe=train_df, x_col="path", y_col="is_Freddie",
                                                    target_size=(256, 256), batch_size=batch_size, class_mode='binary')
test_generator = test_datagen.flow_from_dataframe(dataframe=test_df, x_col="path", y_col="is_Freddie",
                                                  target_size=(256, 256), batch_size=batch_size, class_mode='binary')

# Train the model
model.fit(train_generator, epochs=epochs)

# Evaluate the model on the test set
score = model.evaluate(test_generator)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
