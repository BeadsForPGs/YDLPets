```python
sns.set(style="ticks")
g = sns.jointplot(data=titanic, x='fare', y='age', kind='scatter')

g.fig.suptitle('Titanic Dataset: Age vs. Fare', y=1.02)
g.ax_joint.set_xlabel('Fare')
g.ax_joint.set_ylabel('Age')
g.ax_marg_x.set_title('Fare Distribution')
g.ax_marg_y.set_title('Age Distribution')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'titanic' containing the Titanic dataset

# Count the number of occurrences of each sex
sex_count = titanic['sex'].value_counts()

# Create the bar plot
sns.barplot(x=sex_count.index, y=sex_count.values)

# Set labels and title
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex vs Count in Titanic Dataset')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'titanic' containing the Titanic dataset

# Filter the DataFrame by sex
male_data = titanic[titanic['sex'] == 'male']
female_data = titanic[titanic['sex'] == 'female']

# Set the figure size
plt.figure(figsize=(10, 5))

# Create the bar chart for males
plt.subplot(1, 2, 1)
sns.histplot(data=male_data, x='age', bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution - Male')

# Create the bar chart for females
plt.subplot(1, 2, 2)
sns.histplot(data=female_data, x='age', bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution - Female')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
